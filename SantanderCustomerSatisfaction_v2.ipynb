### Divisão em dados de treino / teste

```python
# Importando a biblioteca necessária
from sklearn.model_selection import train_test_split
```

```python
# Dividindo os dados em treino e teste
X_treino, X_teste, y_treino, y_teste = train_test_split(X, 
                                                        y, 
                                                        test_size = 0.3, 
                                                        random_state = 1000,
                                                        stratify = df.TARGET)
```

```python
# Resultado da divisão dos dados
print(X_treino.shape, X_teste.shape, y_treino.shape, y_teste.shape)
```

### Balanceamento de Classes

```python
# Variável Target está desbalanceada, ou seja,
# Existem mais exemplos de clientes satisfeitos (0)
# a clientes não satisfeitos com o banco (1)

y_treino.value_counts()
```

```python
# Para balancear as classes, será utilizado o método over_sampling
# que cria dados sintéticos para equilibrar as quantidades de 0 e 1

import imblearn
from imblearn.over_sampling import SMOTE
```

```python
over_sampler = SMOTE(k_neighbors = 2)
```

```python
X_btreino, y_btreino = over_sampler.fit_resample(X_treino, y_treino)
```

```python
# Classes balanceadas
y_btreino.value_counts()
```

### Algoritmo KNN

```python
# Lista de possíveis valores de K
vizinhos = list(range(1, 20, 2))
```

```python
# Lista para os scores
cv_scores = []
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score

# Validação cruzada para determinar o melhor valor de k
for k in vizinhos:
    knn = KNeighborsClassifier(n_neighbors = k)
    scores = cross_val_score(knn, X_btreino, y_btreino, cv = 5, scoring = 'accuracy')
    cv_scores.append(scores.mean()) 
```

```python
# Ajustando o erro de classificação
erro = [1 - x for x in cv_scores]
```

```python
# Determinando o melhor valor de k (com menor erro)
optimal_k = vizinhos[erro.index(min(erro))]
print('O valor ideal de k é %d' % optimal_k)
```

```python
# Criando o modelo
modelo_knn = KNeighborsClassifier(n_neighbors = optimal_k)
```

```python
# Treinando o modelo
modelo_knn.fit(X_btreino, y_btreino)
```

```python
import joblib

# Salva o modelo em disco
with open('modelo_knn_v2.pkl', 'wb') as pickle_file:
      joblib.dump(modelo_knn, 'modelo_knn_v2.pkl') 
```

```python
# Previsões
y_pred_knn = modelo_knn.predict(X_teste)
```

```python
from sklearn.metrics import roc_curve, auc, roc_auc_score, confusion_matrix
from sklearn.metrics import accuracy_score

# Matriz de Confusão
confusion_matrix(y_teste, y_pred_knn)
```

```python
# Previsão de probabilidade da classe positiva
y_pred_proba_knn = modelo_knn.predict_proba(X_teste)[:,1]
```

```python
# Curva ROC nos dados e previsões em teste
roc_auc_knn = roc_auc_score(y_teste, y_pred_knn)
print(roc_auc_knn)
```

```python
# Curva ROC nos dados e previsões em teste
fpr_knn, tpr_knn, thresholds = roc_curve(y_teste, y_pred_proba_knn)
```

```python
# AUC em teste
auc_knn = auc(fpr_knn, tpr_knn)
print(auc_knn)
```

```python
# Acurácia em teste
acuracia_knn = accuracy_score(y_teste, y_pred_knn)
print(acuracia_knn)
```
